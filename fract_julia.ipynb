```python
#@title Fractais com conjunto Julia

%pylab inline
import matplotlib.pylab as plt
import numpy as np
from numpy import newaxis
np.warnings.filterwarnings("ignore")

def Julia(z, c ,mxint = 200):
    nint = 0
    while abs(z) <= 4 and nint < 200:
        nint += 1
        z = z * z + c
    return nint
           
def GeraFractal(c, step = 0.005):    
    m = [ [Julia(complex(i,j),c) for j in np.arange(1.0,-1.0,-step)] for i in np.arange(-1.5,1.5, step) ]
    fig = plt.figure(figsize=(15,15))
    plt.imshow(m, cmap="twilight_shifted", extent=[-2,2,-2,2])

GeraFractal(-0.06+0.67j, 0.005)
```

```python
#@title Conjunto Júlia c = -1

JuliaV = np.vectorize(Julia)

def juliafrac(c,x0=-2.0,xf=2.0,y0=-1.5,yf=1.5):
    x = np. linspace(x0,xf, 1000)
    y = np. linspace(y0,yf, 1000)

    # Matriz complexa
    mc = x[:,newaxis ] + 1j*y[newaxis ,:]

    fig = plt.figure(figsize=(10,8))
    plt.imshow(JuliaV(mc, c, mxint = 10).T, extent =[x0,xf,y0,yf],cmap="twilight_shifted") ## cmap= "twilight_shifted" "RdGy")
    plt.show;

juliafrac(-1)
```

```python
#@title Conjunto de Mandelbrot z = z*z +c


def is_stable(c, num_iterations):
    z = 0
    nint = 0
    while abs(z) <= 2 and nint < num_iterations:
        nint += 1
        z = z ** 2 + c
    return nint

is_stableV = np.vectorize(is_stable)    


def Mandelbrotfrac(x0=-2.0,xf=2.0,y0=-1.5,yf=1.5):
    x = np. linspace (x0,xf, 1000)
    y = np. linspace (y0,yf, 1000)

    # Matriz complexa
    mc = x[:,newaxis ] + 1j*y[newaxis ,:]

    plt.figure(figsize=(10,8))
    plt.imshow(is_stableV(mc, 200).T, extent =[x0,xf,y0,yf],cmap="twilight_shifted",interpolation='nearest')
    plt.show;

Mandelbrotfrac()
```

```python
#@title Conjunto de Mandelbrot z = z^3 +c


def is_stable3(c, num_iterations):
    z = 0
    nint = 0
    while abs(z) <= 2 and nint < num_iterations:
        nint += 1
        z = z ** 3 + c
    return nint

is_stableV3 = np.vectorize(is_stable3)    


def Mandelbrotfrac3(x0=-2.0,xf=2.0,y0=-1.5,yf=1.5):
    x = np. linspace (x0,xf, 1000)
    y = np. linspace (y0,yf, 1000)

    # Matriz complexa
    mc = x[:,newaxis ] + 1j*y[newaxis ,:]

    fig = plt.figure(figsize=(10,8))
    plt.imshow(is_stableV3(mc, 200).T, extent =[x0,xf,y0,yf],cmap="twilight_shifted")
    plt.show;

Mandelbrotfrac3()
```
