1.   **Encontre os pontos críticos da função $f(x,y)=x\cdot y$ sujeita a restrição $x^2+y^2=1$:<br>**

Como $x^2+y^2=1$.Substituindo na função objetivo, temos:<br>

$f(x,y)=f(x,\sqrt{1-x^2})=x \cdot \sqrt{1-x^2}$

Assim, temos a seguinte função de uma variável a ser maximizada: $f(x)=x \cdot \sqrt{1-x^2}$ <br>
Calculando a sua derivada pela regra da derivada do produto, temos: 
$$ f'(x) = 1 ⋅ \sqrt{1-x^2} + x ⋅{\frac{1}2} {\frac{1}{\sqrt{1-x^2}}}⋅-2x \\
f'(x) = \frac{1 - 2 x^2}{\sqrt{1-x^2}} $$

Esta função têm seus pontos críticos onde:
 $f'(x) = 0 ⇒  1-2x^2 = 0 ⇒ x = \frac{1}{\sqrt{2}} \text{ e } x = -\frac{1}{\sqrt{2}}$

 Para $x > \frac{1}{\sqrt{2}}$, $f'(x)<0$ ou seja, esté é um ponto de máximo, já para $-\frac{1}{\sqrt{2}} < x < \frac{1}{\sqrt{2}}$, $f'(x)>0$ ou seja, este é um ponto de mínimo.

```python
#@title GRÁFICOS DE $z=x\cdot y$ sujeita a restrição $x^2+y^2=1$
import numpy as np
import matplotlib.pyplot as plt

nx, ny = (50, 50)

x = np.linspace(-1, 1, nx)
y = np.linspace(-1, 1, ny)

xv, yv = np.meshgrid(x, y, indexing='ij')

zz = xv*yv

h = plt.contourf(x, y, zz)
plt.axis('scaled')
plt.colorbar()
plt.show()

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(xv, yv, zz,
                cmap='viridis');
ax.view_init(45, 75);
```

2.   **Calcule as derivadas de segubda ordem da função $ln(1+e^{xy})$: <br>**

- Derivadas parciais de primeira ordem:

$$\dfrac{\partial f}{\partial x} = \frac{1}{1+e^{xy}} ⋅ ye^{xy}  \:\text{ e }\:\dfrac{\partial f}{\partial y} = \frac{1}{1+e^{xy}} ⋅ xe^{xy}$$

$$\dfrac{\partial^2 f}{\partial x^2} = y⋅ \dfrac{\partial }{\partial x}({\frac{e^{xy}}{1+e^{xy}} }) = y ⋅ (\frac{ye^{xy} ⋅ (1+e^{xy}) - e^{xy} ⋅ ye^{xy}}{(1+e^{xy})^2}) = \frac{y^2e^{xy}}{(1+e^{xy})} $$ 

Usando os mesmos princípios:

$$\dfrac{\partial^2 f}{\partial xy} = \dfrac{\partial^2 f}{\partial yx} = \dfrac{\partial }{\partial y}({\frac{ye^{xy}}{1+e^{xy}} }) = \frac{e^{xy} (x y +e^{xy} + 1)}{(e^{xy} + 1)^2}$$

$$\dfrac{\partial^2 f}{\partial y^2} = x⋅ \dfrac{\partial }{\partial y}({\frac{e^{xy}}{1+e^{xy}} }) = \frac{x^2e^{xy}}{(1+e^{xy})} $$

```python
#@title GRÁFICO DE $ln(1+e^{xy})$

zz = np.log(1+np.exp(xv*yv))

h = plt.contourf(x, y, zz)
plt.axis('scaled')
plt.colorbar()
plt.show()

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(xv, yv, zz,
                cmap='viridis');
ax.view_init(45, 75);
```

3.   **Considere a função  $f(x,y)=x^3+y^3-3x-3y+4$. Encontre seus pontos críticos e a natureza de cada um: <br>**

$f(x,y)=x^3+y^3-3x-3y+4$

$$\dfrac{\partial f}{\partial x} = 3x^2-3 ⇒ x = 1 \:\text{e} \: x = -1$$
$$\dfrac{\partial f}{\partial y} = 3y^2-3  ⇒ y = 1 \:\text{e} \: y = -1$$

$$\dfrac{\partial^2 f}{\partial x^2} = 6x ⇒ \dfrac{\partial^2 f(1,y)}{\partial x^2} = 6 \:\text{e} \: \dfrac{\partial^2 f(-1,y)}{\partial x^2} = -6 $$

$$\dfrac{\partial^2 f}{\partial y^2} = 6y ⇒ \dfrac{\partial^2 f(x,1)}{\partial y^2} = 6 \:\text{e} \: \dfrac{\partial^2 f(x,-1)}{\partial y^2} = -6 $$

$$\dfrac{\partial^2 f}{\partial x \partial y} = \dfrac{\partial^2 f}{\partial y \partial x} = 0$$

Calculando o determinante (hessiano da função) de cada ponto:

$$
H(x,y)=\begin{bmatrix}\frac{\partial^{2}f}{\partial x^{2}} & \frac{\partial^{2}f}{\partial x\partial y} \\
\frac{\partial^{2}f}{\partial y\partial x} & \frac{\partial^{2}f}{\partial y^{2}}
\end{bmatrix} = \begin{bmatrix}6x & 0 \\
0 & 6y
\end{bmatrix} = 36xy$$

Para o ponto $(x,y)=(1,1)$:
> $$ H(1,1)=\begin{bmatrix}6 & 0 \\ 0 & 6 \end{bmatrix} = 36 $$
> Como $ \dfrac{\partial^2 f(1,1)}{\partial x^2} >0 \: \text{e} \: H(1,1)>0  ⇒$ > então (1,1) é um ponto de mínimo local de f


Para o ponto $(x,y)=(-1,-1)$:
> Como $ \dfrac{\partial^2 f(-1,-1)}{\partial x^2} < 0 \: \text{e} \: H(-1,-1)
> 0  ⇒$ > então (-1,-1) é um ponto de máximo local de f

Para o ponto $(x,y)=(-1,1) e (x,y)=(1,-1)$:
> $H(-1,1) = -36 <0 $ não é extremante local. Nesse caso os pontos são pontos de sela.


```python
#@title GRÁFICO DE $f(x,y)=x^3+y^3-3x-3y+4$

import warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning) 

nx, ny = (50, 50)

x = np.linspace(-3, 3, nx)
y = np.linspace(-3, 3, ny)

xv, yv = np.meshgrid(x, y, indexing='ij')

zz = xv**3+yv**3-3*xv-3*yv+4

h = plt.contourf(x, y, zz)
plt.axis('scaled')
plt.colorbar()
plt.show()

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.contour3D(xv, yv, zz,3*nx,
                cmap='viridis');
```
