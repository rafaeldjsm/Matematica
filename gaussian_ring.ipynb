## Fundamentação

Teorema : Sendo $p$ um primo a equação $x^2 + y^2 = p$ possui solução inteira se, e somente se, $p = 2$ ou $p≡1(\mod4)$.

```python
def ehprimo(n):
    '''
    Função que verifica se um número é primo
    '''
    if n == 0:
        return False
    else:
        for k in range(int(n**0.5)):
            if n % (k+2) == 0:
                return False
    return True


def ehsumquad(n):
    '''
    Função que verifica se um número pode ser representado pela soma do quadrado de outros dois ineiros.
    '''
    if n == 2 or n % 4 == 1:
        return True
    else:
        return False


def a_p_b_ehprimo(n):
    '''
    retorna a e b talque a^2+b^2=n desde que n seja primo e possa ser representado pela soma de dois quadrados.
    '''
    if not (ehprimo(n) and ehsumquad(n)):
        return False
    else:
        lstquad = [k**2 for k in range(n)]
        lstab = []
        for j in lstquad:
            if (n - j) in lstquad:
                lstab.extend([(j,n-j),(-j,n-j),(-j,-n+j),(j,-n+j)])
    return lstab
```

```python
import matplotlib.pyplot as plt


resfilt0 = []
resfilt = []

plt.figure(figsize=(15,15))
for k in range(1500):
    resp = a_p_b_ehprimo(k)
    if resp:
        resfilt = list(filter(lambda x :(x[0]**2+x[1]**2) <= (1000**2),resp))
    resfilt0.extend(resfilt)

plt.scatter(*zip(*resfilt0), color = 'black', marker = 's', s = 100)
plt.show()
```
