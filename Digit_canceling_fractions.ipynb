The fraction  is a curious fraction. An inexperienced mathematician while attempting to simplify it may incorrectly believe that  is obtained by cancelling the s.

We shall consider fractions like, , to be trivial examples.

Which means fractions where trailing 0's are cancelled are trivial. So we will ignore all the cases where we have to cancel 0's.

You will be given 2 integers  and .  represents the number of digits in Numerator and Denominator, and  represents the exact number of digits to be "cancelled" from Numerator and Denominator. Find every non-trivial fraction, (1) where numerator is less than denominator, (2) and the value of the reduced fraction is equal to the original fraction.

Sum all the Numerators and the Denominators of the original fractions, and print them separated by a space.

Input Format

Input contains two integers  

Constraints



Output Format

Display 2 space separated integers that denote the sum of the Numerators and the sum of the Denominators respectively of original fractions.

```python
n = 3
k = 1

def ehsubst(s0,s1):
    '''
    Verifica se s1 é um combinação com repetição de s0
    '''
    return not bool([k for k in s1 if s1.count(k) > s0.count(k)])

def remol(l1,l2):
    '''
    recebe listas e as subtrai
    '''
    l1t, l2t = l1.copy(), l2.copy()
    for k in l2:
        if k in l1:
            l1t.remove(k)
    return l1t

def arrnj(st,k = 2):
    '''
    recebe uma string e cria subconjuntos, permitindo duplicadas da string original
    retorn uma lista de strings
    '''
    st_inn = [k for k in st]
    while len(st_inn[0]) < k:
        st_inn = [i+j for j in st for i in st_inn if ehsubst(st,(i+j))]
        
    return st_inn


lres = []

for num in range(10**(n-1),10**n):
    # Exclui fracoes triviais multiplas de 10
    for den in range(10**(n-1),num):

        stn = str(num)
        std = str(den)

        dig_com = [i for i in std if (i in stn and i != '0')]


        if k > 1 and len(dig_com) > k:
            dig_com = list(set(arrnj("".join(dig_com),k = k)))

        if dig_com and len(dig_com[0]) >= k:
            for j in dig_com:
                new_n = stn
                new_d = std
                for l in j:
                    new_n = remol(list(new_n),[l])
                    new_d = remol(list(new_d),[l])

                if (len(new_d) == len(stn) - k) and (len(new_n) == len(new_d)):
                    new_d = int("".join(new_d))
                    new_n = int("".join(new_n))
                    
                    if (num * new_d == new_n * den) and new_d*new_n != 0:
                        print(num,den, dig_com, new_n, new_d)
                        l_item = [den,num,new_d,new_n]
                        if l_item not in lres:
                            lres.append(l_item)


print(sum([k[0] for k in lres]),sum([k[1] for k in lres]))
```
