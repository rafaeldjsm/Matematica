```python
import matplotlib.pyplot as plt
import numpy as np

N = 10

mseh = np.array([[i,j,i&j] for i in range(1,N+1) for j in range(1,N+1)])
plt.subplots(figsize=(20,20))
ax = plt.axes(projection='3d')
ax.plot3D(mseh[:,0], mseh[:,1], mseh[:,2], 'gray')
plt.show()
```

```python
def f(x, y):
    return x & y 

x = [*range(1,11)]
y = [*range(1,11)]

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(20,20))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```

```python
import pandas as pd
import seaborn as sns

data = pd.DataFrame({'X': mseh[:,0], 'Y': mseh[:,1], 'Z': mseh[:,2]})
data_pivoted = data.pivot("Y","X", "Z")

plt.subplots(figsize=(20,20))

ax = sns.heatmap(data_pivoted)
plt.show()
```

```python
def f(x, y):
    return x & y 

x = [*range(1,20)]
y = [*range(1,20)]

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.subplots(figsize=(20,20))

contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```

```python
def isPowerOfTwo(x):
    # First x in the below expression
    # is for the case when x is 0
    return (x and (not(x & (x - 1))) )

def f(x, y):
    bw = x & y 
    return bw if isPowerOfTwo(bw) else 0

f = np.vectorize(f)

x = [*range(1,30)]
y = [*range(1,30)]

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.subplots(figsize=(20,20))
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```
